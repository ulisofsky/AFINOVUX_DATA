```python
import streamlit.components.v1 as components

def crear_destellos():
    """
    Inyecta JS y CSS para crear destellos animados en coordenadas aleatorias.
    Los destellos serán dorados para combinar con el tema del templo solar.
    """
    components.html(
        """
        <style>
            /* Estilo del punto de luz */
            .sparkle {
                position: fixed;
                width: 6px;
                height: 6px;
                border-radius: 50%;
                background-color: #FFD700; /* Dorado */
                box-shadow: 0 0 10px 5px rgba(255, 215, 0, 0.4);
                pointer-events: none; /* Para que no impida hacer clic */
                z-index: 9999; /* Encima de todo */
                animation: sparkle-anim 1.5s ease-out forwards;
            }

            @keyframes sparkle-anim {
                0% {
                    transform: scale(0);
                    opacity: 0;
                }
                50% {
                    transform: scale(1);
                    opacity: 1;
                }
                100% {
                    transform: scale(0);
                    opacity: 0;
                }
            }
        </style>

        <script>
            function createSparkle() {
                // Crear elemento div
                const sparkle = document.createElement('div');
                sparkle.classList.add('sparkle');

                // Coordenadas aleatorias (0% a 100% de la ventana)
                const x = Math.random() * 100;
                const y = Math.random() * 100;

                // Aplicar posición
                sparkle.style.left = x + 'vw';
                sparkle.style.top = y + 'vh';

                // Añadir al cuerpo de la página
                document.body.appendChild(sparkle);

                // Eliminar el elemento después de la animación (1.5s)
                setTimeout(() => {
                    sparkle.remove();
                }, 1500);
            }

            // Crear un destello cada 300ms (ajustable)
            setInterval(createSparkle, 300);
        </script>
        """,
        height=0,  # El componente no ocupa espacio visual
        width=0
    )
```
